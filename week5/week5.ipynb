```python
!pip install deep-phonemizer
import torch
import torchaudio

torch.random.manual_seed(0)
device = "cuda" if torch.cuda.is_available() else "cpu"

print(torch.__version__)
print(torchaudio.__version__)
print(device)
```

```python
import IPython
import matplotlib.pyplot as plt
```

```python
text = "Hi, I'm tae young kim. I'm an undergraduate researcher passionate about AI, ML, and robotics with hands-on project experience. Outside of my studies, I focus on weight training and improving my English conversation skills."
```

```python
#Text를 학습할 수 있는 int값으로 변환하는 과정.
symbols = "_-!'(),.:;? abcdefghijklmnopqrstuvwxyz"
look_up = {s: i for i, s in enumerate(symbols)}
symbols = set(symbols)


def text_to_sequence(text):
    text = text.lower()
    return [look_up[s] for s in text if s in symbols]

print(text_to_sequence(text))
```

```python
#Tacotron2로 text를 음소 list로 바꿈
bundle = torchaudio.pipelines.TACOTRON2_WAVERNN_PHONE_LJSPEECH

processor = bundle.get_text_processor()

with torch.inference_mode():
    processed, lengths = processor(text)

print(processed)
print(lengths)
print([processor.tokens[i] for i in processed[0, : lengths[0]]])
```

```python
#음소 숫자 시퀀스를 넣어서 멜 스펙트로그램 생성

# 1. Tacotron2 모델 불러오기
tacotron2 = bundle.get_tacotron2()

# 2. 모델과 데이터를 동일한 장치(CPU 또는 GPU)로 이동
device = "cuda" if torch.cuda.is_available() else "cpu"
tacotron2 = tacotron2.to(device)
processed = processed.to(device)
lengths = lengths.to(device)

# 3. 모델을 사용해 멜 스펙트로그램 생성
print("\nGenerating spectrogram...")
with torch.inference_mode():
    # .infer() 메소드를 사용하여 스펙트로그램 추론
    spec, spec_lengths, _ = tacotron2.infer(processed, lengths)

print("✓ Spectrogram generation successful!")
# 생성된 스펙트로그램의 형태(shape)를 출력해봅니다.
print("Spectrogram shape:", spec.shape)
print("Spectrogram lengths:", spec_lengths)
```

```python
from IPython.display import Audio
```

```python
# --- 옵션 1: 번들에 포함된 고품질 WaveRNN 보코더 사용 ---

print("--- Option 1: Using WaveRNN Vocoder (High Quality) ---")

# 1. 번들에서 WaveRNN 보코더 불러오기
vocoder = bundle.get_vocoder()
vocoder = vocoder.to(device)
# 2. 보코더를 사용해 멜 스펙트로그램을 음성 파형으로 변환
print("Generating waveform with WaveRNN...")
with torch.inference_mode():
    waveform, _ = vocoder(spec, spec_lengths)

print(f"Sample rate: {vocoder.sample_rate}")

# Jupyter Notebook 또는 Colab 환경에서 바로 들어보기
print("Playing WaveRNN audio:")
display(Audio(waveform[0].to("cpu"), rate=vocoder.sample_rate))
```

```python
# --- 옵션 2: 비교를 위한 빠른 Griffin-Lim 보코더 사용 ---

print("\n--- Option 2: Using Griffin-Lim Vocoder (Fast, Lower Quality) ---")

bundle = torchaudio.pipelines.TACOTRON2_GRIFFINLIM_PHONE_LJSPEECH

processor = bundle.get_text_processor()
tacotron2 = bundle.get_tacotron2().to(device)
vocoder = bundle.get_vocoder().to(device)

with torch.inference_mode():
    processed, lengths = processor(text)
    processed = processed.to(device)
    lengths = lengths.to(device)
    spec, spec_lengths, _ = tacotron2.infer(processed, lengths)
waveforms, lengths = vocoder(spec, spec_lengths)

# Jupyter Notebook 또는 Colab 환경에서 바로 들어보기
print("Playing Griffin-Lim audio:")
display(Audio(waveforms[0].to("cpu"), rate=vocoder.sample_rate))
```

```python
#Waveglow Vocoder
waveglow = torch.hub.load(
    "NVIDIA/DeepLearningExamples:torchhub",
    "nvidia_waveglow",
    model_math="fp32",
    pretrained=False,
)
checkpoint = torch.hub.load_state_dict_from_url(
    "https://api.ngc.nvidia.com/v2/models/nvidia/waveglowpyt_fp32/versions/1/files/nvidia_waveglowpyt_fp32_20190306.pth",  # noqa: E501
    progress=False,
    map_location=device,
)
state_dict = {key.replace("module.", ""): value for key, value in checkpoint["state_dict"].items()}

waveglow.load_state_dict(state_dict)
waveglow = waveglow.remove_weightnorm(waveglow)
waveglow = waveglow.to(device)
waveglow.eval()

with torch.no_grad():
    waveform = waveglow.infer(spec)

print("Playing Waveglow audio:")
display(Audio(waveform[0].to("cpu"), rate=vocoder.sample_rate))
```
